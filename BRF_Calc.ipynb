```python
interval = '1d'
period1 = int(time.mktime(datetime.datetime(2016, 7, 1, 23, 59).timetuple()))
period2 = int(time.mktime(datetime.datetime(2021, 7, 1, 23, 59).timetuple()))
ss_data = pd.DataFrame()

xlwriter = pd.ExcelWriter('historical prices.xlsx', engine='openpyxl')
for ticker in ss_tickers:
    try:
        query_string = f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}?period1={period1}&period2={period2}&interval={interval}&events=history&includeAdjustedClose=true'
        df = pd.read_csv(query_string)
        df.drop(columns=['Open', 'Close', 'Low', 'High', 'Volume'], inplace=True)
        df['Date'] = pd.to_datetime(df['Date'])
        df.set_index('Date', inplace=True)
        df.rename(columns={'Adj Close': ticker}, inplace=True)
        ss_data = pd.concat([ss_data, df], axis=1)
    except:
        print('No data: ', ticker)
```

```python
new_factor['small_RMS'] = new_factor['Small Responsible'] - new_factor['Small Secular']
new_factor['big_RMS'] = new_factor['Big Responsible'] - new_factor['Big Secular']
```

```python
new_factor['SMB_RMS'] = 1/2 * (new_factor['Small Responsible'] + new_factor['Small Secular']) - 1/2 * (new_factor['Big Responsible'] + new_factor['Big Secular'])
```

```python
new_factor
```

```python
new_factor['BRF'] = 1/2 * (new_factor['Small Responsible'] + new_factor['Big Responsible']) - 1/2 * (new_factor['Small Secular'] - new_factor['Big Secular'])
```

```python
new_factor['SMB_Final'] = 1/2 * (new_factor['SMB_BM'] + new_factor['SMB_RMS'])
```
