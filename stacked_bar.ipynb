```python
import pandas as pd 
import time
import datetime
test_list = pd.read_csv(r'C:\Users\44738\Desktop\Desktop Files\Dissertation\test_stock_list.csv', encoding='latin1')


# symbols = list(table[0]['Symbol'])

# for symbol in symbols[0:5]:
#     print(symbol)
symbols = []
temp = list(test_list['stock'])
temp = ['^GSPC']
for i in temp:
    j = i.lstrip()
    symbols.append(j)
```

```python
interval = '1d'
period1 = int(time.mktime(datetime.datetime(2016, 7, 1, 23, 59).timetuple()))
period2 = int(time.mktime(datetime.datetime(2021, 7, 1, 23, 59).timetuple()))
bs_data = pd.DataFrame()

xlwriter = pd.ExcelWriter('historical prices.xlsx', engine='openpyxl')
for ticker in symbols:
    try:
        query_string = f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}?period1={period1}&period2={period2}&interval={interval}&events=history&includeAdjustedClose=true'
        df = pd.read_csv(query_string)
        df.drop(columns=['Open', 'Close', 'Low', 'High', 'Volume'], inplace=True)
        df['Date'] = pd.to_datetime(df['Date'])
        df.set_index('Date', inplace=True)
        df.rename(columns={'Adj Close': ticker}, inplace=True)
        bs_data = pd.concat([bs_data, df], axis=1)
    except:
        print('No data for: ', ticker)
```

```python
bs_data
```

```python
bs_returns = bs_data.pct_change()[1:]
weights = [1/len(bs_data.columns)] * len(bs_data.columns)

bs_weighted = (weights * bs_returns)
bs_weighted
```

```python
bs_portfolio = pd.DataFrame(bs_weighted.sum(axis=1)).rename(columns={0:'Screened_SP500'})
bs_portfolio
```
